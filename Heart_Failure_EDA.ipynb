```python
#Range and average values of various attributes of people who survived
death_data = dataset[dataset['death'] == 0]
attributes = ['age', 'creatinine_phosphokinase', 'ejection_fraction', 'platelets', 'serum_creatinine', 'serum_sodium']

for a in attributes:
    a_min = death_data[a].min()
    a_max = death_data[a].max()
    a_avg = death_data[a].mean()
    print("The range of {} for patients who survived is {}-{}".format(a, a_min, a_max))
    print("The average {} for the same is {:.2f}".format(a, a_avg))
```

```python
#Range and average values of various attributes of people who died
survived_data = dataset[dataset['death'] == 1]
attributes = ['age', 'creatinine_phosphokinase', 'ejection_fraction', 'platelets', 'serum_creatinine', 'serum_sodium']

for a in attributes:
    a_min = survived_data[a].min()
    a_max = survived_data[a].max()
    a_avg = survived_data[a].mean()
    print("The range of {} for patients who died is {}-{}".format(a, a_min, a_max))
    print("The average {} for the same is {:.2f}".format(a, a_avg))
```
